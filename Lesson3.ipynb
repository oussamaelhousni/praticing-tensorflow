```python
import numpy as np
# create the dataset
from sklearn.datasets import make_circles
n_samples = 1000
X,y = make_circles(n_samples=n_samples, noise=0.15)

X[:10], y[:10]
```

```python
# visualize the data
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.scatter(X[:, 0],X[:,1],c=y)
plt.show()
```

```python
# create the model
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(50,activation="relu"),
    tf.keras.layers.Dense(20,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

# compile the model
model.compile(
    loss="binary_crossentropy",
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"])

# fit the model
history = model.fit(X,y,epochs=1000)
```

```python
model.predict(X)
```

```python
import pandas as pd
def plot_learning_curve(history):
    learning_df = pd.DataFrame(history.history)
    learning_df.plot()
    learning_df.head()
    plt.xlabel("epoches")
    plt.ylabel("loss")
```

```python
plot_learning_curve(history)
```

```python
def plot_decision_boundary(model,X,y):
    # define the axis  boundries
    x_min, x_max = X[:,0].min() - 0.1,  X[:,0].max() + 0.1
    y_min, y_max = X[:,1].min() - 0.1,  X[:,1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))
    # create X value
    X_in = np.c_[xx.ravel(),yy.ravel()]
    y_pred = model.predict(X_in)
    if len(y_pred[0]) > 1:
        # for multiclass classification
        y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        y_pred = np.round(y_pred).reshape(xx.shape)
        plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
        plt.scatter(X[:, 0],X[:,1],c=y,cmap=plt.cm.RdYlBu,alpha=0.7)
        plt.xlim(xx.min(),xx.max())
        plt.ylim(yy.min(),yy.max())
        plt.show()
        
```

### Using  callbacks
Tensorflow callbacks are functions or blocks of code which are executed during a specific instant while training a Deep Learning Model.

*LearningRateScheduler* : At the beginning of every epoch, this callback gets the updated learning rate value from schedule function provided at __init__, with the current epoch and current learning rate, and applies the updated learning rate on the optimizer.

Arguments

schedule: a function that takes an epoch index (integer, indexed from 0) and current learning rate (float) as inputs and returns a new learning rate as output (float).
verbose: int. 0: quiet, 1: update messages.

```python
lr_schedular = tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-4 * 10**(epoch/20))
```
